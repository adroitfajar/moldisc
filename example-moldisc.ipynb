# MolDisc Tutorial

This tutorial provides a basic introduction to using the MolDisc for molecualr generation. 

## Setup and installation

Download the MolDisc from the github repository https://github.com/adroitfajar/moldisc and dollow the setup instructions.

## Importing libraries and setting the dataset

At the begining of the python file import following libraries.

```python
import time
from moldisc import *
```

## Specifying parameters for MolDisc

### Specify a project name

A unique project name has to be specified for each moelcular generation experiment. 

```python
project_name = "data_1"
```

At that of start of execution of MolDisc a folder with the project name will be created inside the ./projects folder. Further folders with names ./SMILESX, ./GPT, ./tmp, and ./final are created inside the project folder. 



### Specify data

Create a folder with the same name as the project name inside the ./data folder and copy the labelled and unlabelled data. Data has to be in a .csv file. Labelled data should must be names as **labeled.csv** with columns 'SMILES' and 'property' with SMILES text and numerical property values respectively. The unlabelled data only consists of SMILES in a file named as **labelled.csv**.   

### Parameters for SMILES-X

Following parameters specify the hyperparameters for SMILES-X

```python
# Defining the hyperparameters bounds for SMILESX 
embed_bounds = [8, 16, 32, 64, 128, 256, 512] # embedding size
lstm_bounds = [8, 16, 32, 64, 128, 256, 512] # number of units in the LSTM layer
tdense_bounds = [8, 16, 32, 64, 128, 256, 512] # number of units in the dense layer
bs_bounds = [8, 16, 32, 64, 128, 256, 512] # batch size
lr_bounds = [2., 2.5, 3., 3.5, 4.0, 4.5] # learning rate
```

### Parameters for GPT and molecualr generation

Following parameters specify the training condition for GPT model for molecular generations. 

## Running the MolDisc

```python
moldisc(project_name = "data_1",
         input_file_labeled = 'labeled.csv',
         input_file_unlabeled = 'unlabeled.csv',
         project_folder = './projects', 
          data_err=None, # Error on the property
          data_name='test',
          data_units='units',
          data_label='Test label',
          smiles_concat=True,
          #outdir =  '/outputs/' + str(0) +"/",
          geomopt_mode='off', # Zero-cost geometry optimization
          bayopt_mode='off', # Bayesian optimization
          train_mode='on', # Train
          model_type = 'regression', # 'regression' or 'classification'
          scale_output = True,
          bs_bounds=bs_bounds,
          lr_bounds=lr_bounds,
          embed_bounds=embed_bounds,
          lstm_bounds=lstm_bounds,
          tdense_bounds=tdense_bounds,
          k_fold_number=2, # Number of cross-validation splits
          n_runs=5 ,# Number of runs per fold
          check_smiles=True, # Verify SMILES validity via RDKit
          augmentation=True, # Augment the data or not
          bayopt_n_rounds=5,
          bayopt_n_epochs=25,
          bayopt_n_runs=25,
          n_gpus=1,
          n_epochs=3,
          log_verbose=True, # To send print outs both to the file and console
          train_verbose=True,
          gpt_pretrained_model='gpt2', 
          gpt_augmentation=2,
          gpt_data_split=0.8,
          #gpt_output_dir ="./output/",
          gpt_tr_epochs=100,
          gpt_tr_batch_size=2,
          gpt_eval_batch_size=2,
          gpt_warmup_steps=10,
          gpt_decay =0.01,
          #gpt_log_dir ="./log/",
          gpt_patience= 25,
          gpt_device = 2,
          gpt_num_generation=10000,
          gpt_remove_ionic = 'all',
          gpt_num_attempts =5,
          max_generation=100,
          cycles = 1,
          sa_score=5,
          cutoff=0.5,
          option_save=True,
          option_show=True,
          num_mol_top=5,
          num_mol_bottom=5,
          cuda=0,
          patience=100,
          remove_tmp=True)
```
